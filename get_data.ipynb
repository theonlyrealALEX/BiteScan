```python
import requests
import os

# Define the path to the file containing image URLs
urls_path = './data/urls_images/urls_tick_google.txt'

# Define the path to the output directory for images
output_dir = './data/raw/tick/'

# Open the file containing URLs, read it, strip unnecessary whitespace,
# split it into rows and initialize the total images counter
with open(urls_path, 'r') as f:
    rows = f.read().strip().split("\n")

total = 0

# Make sure the output directory exists; if not, create it
os.makedirs(output_dir, exist_ok=True)

# Download each image
for url in rows:
    try:
        # Try to download the image
        r = requests.get(url, timeout=60)

        # Save the image to disk
        p = os.path.sep.join([output_dir, "{}.jpg".format(str(total).zfill(8))])
        with open(p, "wb") as img_file:
            img_file.write(r.content)

        # Update the counter
        print("[INFO] downloaded: {}".format(p))
        total += 1

    # Handle if any exceptions are thrown during the download process
    except Exception as e:
        print("[INFO] error downloading {}...skipping".format(url))
        print("[INFO] error message: {}".format(e))
```
